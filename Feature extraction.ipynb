```python
print("Plotting Time Variable")
print("======================")

# Assuming you have a time-related column named 'Effective To Date'
# Convert the 'Effective To Date' to datetime format
customer_df['Effective To Date'] = pd.to_datetime(customer_df['Effective To Date'])

# Extract year, month, and day information
customer_df['year'] = customer_df['Effective To Date'].dt.year
customer_df['month'] = customer_df['Effective To Date'].dt.month
customer_df['day'] = customer_df['Effective To Date'].dt.day

# Plot the extracted time-related features
plt.figure(figsize=(12, 6))

# Plot the year distribution
plt.subplot(131)
sns.countplot(data=customer_df, x='year')
plt.title("Year Distribution")

# Plot the month distribution
plt.subplot(132)
sns.countplot(data=customer_df, x='month')
plt.title("Month Distribution")

# Plot the day distribution
plt.subplot(133)
sns.countplot(data=customer_df, x='day')
plt.title("Day Distribution")

plt.tight_layout()
plt.show()
```
